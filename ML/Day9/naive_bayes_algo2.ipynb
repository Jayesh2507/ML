```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv(r'Z:\ML\files\NB_loan_data_set.csv',usecols=['Gender','Married','Education','Self_Employed','Property_Area','Loan_Status'])
df
```

```python
df = df.dropna()
df
```

```python
input = df.drop('Loan_Status',axis='columns')
target = df['Loan_Status']
target
```

```python
numeric = LabelEncoder()
```

```python
input['gender_new'] = numeric.fit_transform(input['Gender'])
input['married_new'] = numeric.fit_transform(input['Married'])
input['education_new'] = numeric.fit_transform(input['Education'])
input['self_employed_new'] = numeric.fit_transform(input['Self_Employed'])
input['property_area_new'] = numeric.fit_transform(input['Property_Area'])
input
```

```python
input = input.drop(['Gender','Married','Education','Self_Employed','Property_Area'],axis='columns')
input
```

```python
classifier = GaussianNB()
classifier.fit(input,target)
```

```python
classifier.score(input,target)
```
