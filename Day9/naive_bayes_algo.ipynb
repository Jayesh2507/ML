```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv(r'Z:\ML\files\playsheet_dataset.csv')
df
```

```python
#removing output column from the dataframe and storing it into target variable
input = df.drop('Play',axis='columns')
target = df['Play']
```

```python
#creating object of LabelEncoder class 
Numeric = LabelEncoder()
```

```python
#converting input dataframe into numeric values because every algorithm works on numeric data.
input['outlook_new'] = Numeric.fit_transform(input['Outlook'])
input['temp_new'] = Numeric.fit_transform(input['Temp'])
input['humidity_new'] = Numeric.fit_transform(input['Humidity'])
input['windy_new'] = Numeric.fit_transform(input['Windy'])
input
```

```python
# now removing the text data from input dataframe
input = input.drop(['Outlook','Temp','Humidity','Windy'],axis='columns')
input
```

```python
#applying the guassian naive bayes algorithm
classifier = GaussianNB()
classifier.fit(input,target)
```

```python
# checking accuracy
classifier.score(input,target)
```

```python
#predictiong for new row
classifier.predict([[1,1,0,0]])
```
